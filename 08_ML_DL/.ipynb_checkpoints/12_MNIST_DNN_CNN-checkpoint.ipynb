```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import tensorflow.keras.utils as utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import numpy as np
```

# DNN

```python
# ▶ 1. 데이터 셋 생성
width = 28; height = 28
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(60000, width * height).astype('float32') / 255.0
x_test = x_test.reshape(10000, width * height).astype('float32') / 255.0

# 데이터 분리
x_val = x_train[50000:]
y_val = y_train[50000:]

x_train = x_train[:50000]
y_train = y_train[:50000]

# 원-핫 인코딩
y_train = utils.to_categorical(y_train)
y_val = utils.to_categorical(y_val)
y_test = utils.to_categorical(y_test)
```

```python
plt_row = 5; plt_col = 5
plt.rcParams['figure.figsize'] = (16, 9)
plt.rc('font', family = 'Malgun Gothic')
plt.subplots_adjust(hspace = 0.4,
                    wspace = 0.3)

fig, axarr = plt.subplots(plt_row, plt_col)
for i in range(plt_row * plt_col):
    sub_plot = axarr[ (i//5), (i%5)]
    sub_plot.imshow(x_test[i].reshape(width, height))
    sub_plot.axis('off')
    sub_plot.set_title('실제 :'+str(np.argmax(y_test[i])))
    
plt.show()
```

```python
# ▶ 2. 모델 생성
model = Sequential()
model.add(Dense(256, input_dim = width*height, activation = 'relu'))
model.add(Dense(256, input_dim = width*height, activation = 'relu'))
model.add(Dense(128, input_dim = width*height, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))

# ▶ 3. 모델 - 학습과정 설정
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# ▶ 4. 모델 학습
earlyStopping = EarlyStopping(patience = 10)
hist = model.fit(x_train, y_train, epochs = 30, batch_size = 32, validation_data = (x_val, y_val), 
                 callbacks = [earlyStopping])
```

```python
# ▶ 5. 모델 학습과정 표시하기
fig, loss_ax = plt.subplots(figsize = (16, 9))

loss_ax.plot(hist.history['loss'], 'y', label = 'train loss')
loss_ax.plot(hist.history['val_loss'], 'g', label = 'val_loss')

loss_ax.set_xlabel('epochs')
loss_ax.set_ylabel('loss')
loss_ax.legend(loc = 'upper left')

acc_ax = loss_ax.twinx()
acc_ax.plot(hist.history['accuracy'], 'b', label = 'accuracy')
acc_ax.plot(hist.history['val_accuracy'], 'r', label = 'val_accuracy')

acc_ax.set_ylabel('accuracy')
acc_ax.legend(loc = 'lower left')

plt.show()
```

```python
# ▶ 6. 모델 평가 및 성능지표
loss_and_acc = model.evaluate(x_test, y_test)

print('Loss :', round(loss_and_acc[0]*100, 2), '%')
print('Accuracy :', round(loss_and_acc[1]*100, 2), '%')
```

```python
# 모델을 사용하여 예측이 틀린 데이터만 출력
yhat = model.predict(x_test).argmax(axis = 1) # 예측치
y = y_test.argmax(axis = 1) # 실제값

plt_row = 22; plt_col = 10
plt.rcParams['figure.figsize'] = (20, 50)
plt.rc('font', family = 'Malgun Gothic')
plt.subplots_adjust(hspace = 0.4, wspace = 0.3)
fig, axarr = plt.subplots(plt_row, plt_col)

i = 0; cnt = 0 # i = yhat과 y를 액세스 할 index ,cnt = 출력횟수
while ( cnt < (plt_row*plt_col) ) & ( i<len(y) ):
    if yhat[i] == y[i]:
        i += 1
        continue

    sub_plot = axarr[cnt//10, cnt%10]
    sub_plot.imshow(x_test[i].reshape(width, height))
    sub_plot.axis('off')
    title = '실제 :' + str(y[i]) + ', 예측 :' + str(yhat[i])
    sub_plot.set_title(title)
    
    i += 1
    cnt += 1
```

# CNN (합성곱 신경망)

```python
# ▶ 1. 데이터 셋 생성
width = 28; height = 28
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(60000, width, height, 1).astype('float32') / 255.0
x_test = x_test.reshape(10000, width, height, 1).astype('float32') / 255.0

# 데이터 분리
x_val = x_train[50000:]
y_val = y_train[50000:]

x_train = x_train[:50000]
y_train = y_train[:50000]

# 원-핫 인코딩
y_train = utils.to_categorical(y_train)
y_val = utils.to_categorical(y_val)
y_test = utils.to_categorical(y_test)
```

```python
# ▶ 2. 모델 생성
model = Sequential()

model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (28, 28, 1)))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(32, (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Dropout(0.1))
model.add(Flatten())

model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))

# ▶ 3. 모델 - 학습과정 설정
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# ▶ 4. 모델 학습
earlyStopping = EarlyStopping(patience = 10)
hist = model.fit(x_train, y_train, epochs = 30, batch_size = 32, validation_data = (x_val, y_val), 
                 callbacks = [earlyStopping])
```

```python
# ▶ 5. 모델 학습과정 표시하기
fig, loss_ax = plt.subplots(figsize = (16, 9))

loss_ax.plot(hist.history['loss'], 'y', label = 'train loss')
loss_ax.plot(hist.history['val_loss'], 'g', label = 'val_loss')

loss_ax.set_xlabel('epochs')
loss_ax.set_ylabel('loss')
loss_ax.legend(loc = 'upper left')

acc_ax = loss_ax.twinx()
acc_ax.plot(hist.history['accuracy'], 'b', label = 'accuracy')
acc_ax.plot(hist.history['val_accuracy'], 'r', label = 'val_accuracy')

acc_ax.set_ylabel('accuracy')
acc_ax.legend(loc = 'lower left')

plt.show()
```

```python
# ▶ 6. 모델 평가 및 성능지표
loss_and_acc = model.evaluate(x_test, y_test)

print('Loss :', round(loss_and_acc[0]*100, 2), '%')
print('Accuracy :', round(loss_and_acc[1]*100, 2), '%')
```

```python
# 모델을 사용하여 예측이 틀린 데이터만 출력
yhat = model.predict(x_test).argmax(axis = 1) # 예측치
y = y_test.argmax(axis = 1) # 실제값

plt_row = 10; plt_col = 10
plt.rcParams['figure.figsize'] = (20, 20)
plt.rc('font', family = 'Malgun Gothic')
plt.subplots_adjust(hspace = 0.4, wspace = 0.3)
fig, axarr = plt.subplots(plt_row, plt_col)

i = 0; cnt = 0 # i = yhat과 y를 액세스 할 index ,cnt = 출력횟수
while ( cnt < (plt_row*plt_col) ) & ( i<len(y) ):
    if yhat[i] == y[i]:
        i += 1
        continue

    sub_plot = axarr[cnt//10, cnt%10]
    sub_plot.imshow(x_test[i].reshape(width, height))
    sub_plot.axis('off')
    title = '실제 :' + str(y[i]) + ', 예측 :' + str(yhat[i])
    sub_plot.set_title(title)
    
    i += 1
    cnt += 1
```
