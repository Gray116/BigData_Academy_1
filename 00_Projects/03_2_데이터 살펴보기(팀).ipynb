# 평균 점유율 별 승률

```python
df_poss = df1.loc[ : , ['Short Name','average_possession', 'xg_for_avg_overall', 'total_goal_count',
                       'win_percentage', 'leading_at_half_time_percentage']]
```

```python
df_poss = df_poss.sort_values(by='average_possession', ascending=False).head(10).reset_index(drop=True)

df_poss
```

```python
plt.figure(figsize=(16, 9))
sns.scatterplot(data=df_poss, x = 'win_percentage', y = 'average_possession', hue = 'Short Name',
               legend=False)
plt.rcParams['font.family'] = 'MalGun Gothic'
plt.title('상위 10팀의 20-21시즌 비교', fontsize = 15)
plt.suptitle('평균 점유율(average possession)과 승률의(win percentage)의 관계', fontsize = 20)

for n in range(0, len(df_poss)): 
    if (df_poss['average_possession'][n]) & (df_poss['win_percentage'][n]):
        plt.text(df_poss['win_percentage'][n], df_poss['average_possession'][n] + 0.02,
             df_poss['Short Name'][n], fontsize=11, verticalalignment='bottom' , horizontalalignment='center')
    
plt.grid()
plt.show()
```

# 전반 30분 이전, 후반 30분이후 득점이 많은 상위 10팀의 승률

```python
df_win_count = df1.loc[ : , ['Short Name','wins', 'win_percentage', 'leading_at_half_time_percentage',
                            'total_goal_count', 'goals_scored_min_11_to_20', 'goals_scored_min_21_to_30',
                            'goals_scored_min_71_to_80', 'goals_scored_min_81_to_90']]

df_win_count
```

```python
df_win_count['goals_scored_71_to_90'] = df_win_count['goals_scored_min_71_to_80'] + df_win_count['goals_scored_min_81_to_90']
df_win_count['goals_scored_11_to_30'] = df_win_count['goals_scored_min_11_to_20'] + df_win_count['goals_scored_min_21_to_30']

df_win_count
```

```python
df_win_count = df_win_count.sort_values(by='goals_scored_71_to_90', ascending=False).head(10).reset_index(drop=True)

df_win_count
```

```python
df_win_count.drop(columns=['goals_scored_min_11_to_20', 'goals_scored_min_21_to_30', 
                  'goals_scored_min_71_to_80', 'goals_scored_min_81_to_90'], inplace=True)
```

```python
df_win_count
```

```python
df_win_count['11_to_30_GoalP'] = round((df_win_count['goals_scored_11_to_30'] / df_win_count['total_goal_count']) * 100, 1)
df_win_count['71_to_90_GoalP'] = round((df_win_count['goals_scored_71_to_90'] / df_win_count['total_goal_count']) * 100, 1)

df_win_count
```

```python
df_win_count.sort_values(by='win_percentage', ascending=False)
```
