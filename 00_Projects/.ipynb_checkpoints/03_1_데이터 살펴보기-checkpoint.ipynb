## 효율적인 공격을 했던 선수는 누구였을까?
- 내 생각에는 적은 시간에 많은 골을 넣는 선수라고 생각한다.
*  Mins: Minutes played<br>
    Goals: Total goals<br>
    Assists: Total assists<br>
    SpG: Shots per game<br>
    KeyP: Key passes per game<br>
    Drb: Dribbles per game<br>
    Fouled: Fouled per game<br>
    Off: Offsides per game<br>
    Disp: Dispossessed per game<br>
    UnsTch: Bad control per game<br>

```python
# 전체 데이터 중, 공격수만 추출하여 따로 데이터 생성
df_forward = df.loc[df['position'] == 'Forward']
```

```python
df_forward.head()
```

```python
print('appearances_overall' in df_forward.columns)
```

```python
df_forward = df_forward.loc[ : , ['Player','minutes_played_overall', 'appearances_overall', 'Goals', 'SpG']]

df_forward.head()
```

```python
df_forward.rename(columns={'minutes_played_overall' : 'minutes', 'appearances_overall' : 'apps'}, inplace=True)
```

```python
df_forward.sort_values(by='Goals', ascending=False, inplace=True)

df_forward.head(20)
```

```python
# 경기 당 슈팅 수 (SpG)와 경기당 골수(Goals / apps)로 산점도 그려보기
```

```python
Goalspg = round(df_forward['Goals'] / df_forward['apps'], 2)
```
