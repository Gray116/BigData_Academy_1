```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/teams_data/teams_stat.csv')

df
```

```python
df.drop(columns='Unnamed: 0', inplace=True)

df
```

```python
df.sort_values(by='Short Name', inplace=True)
df.reset_index(drop=True)

df
```

```python
print(df.columns.values)
```

# 전반 15분에 득점이 가장 많은 상위 10팀

```python
df_poss = df.loc[ : , ['Short Name','average_possession', 'goals_scored_half_time', 'goals_scored_min_11_to_20']]
```

```python
df_poss = df_poss.sort_values(by='average_possession', ascending=False).head(10)
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(data=df_poss, x = 'average_possession', y = 'goals_scored_min_11_to_20', hue='Short Name')

plt.show()
```
