```python
import pandas as pd # 데이터 프레임화 해야하기 때문
import numpy as np
```

# 포지션별 데이터를 하나로

```python
# ▶ 1. 데이터 셋 불러오기
df1 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/20_21_players/Spain/spa_2021_defensive_stats.csv',
                  encoding = "UTF8").drop(['Unnamed: 0'],axis=1)
df2 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/20_21_players/Spain/spa_2021_offensive_stats.csv',
                  encoding = "UTF8").drop(['Unnamed: 0'],axis=1)
df3 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/20_21_players/Spain/spa_2021_passing_stats.csv', 
                  encoding = "UTF8").drop(['Unnamed: 0'],axis=1)
```

```python
# ▶ 2. 결측치 변경 및 필요없는 열 제거
df1 = df1.replace(np.nan, "0")
df2 = df2.replace(np.nan, "0")
df3 = df3.replace(np.nan, "0")

df1.head()
```

```python
df1.drop(['Rating', 'Drb', 'Offsides'], axis = 1, inplace = True)
df2.drop(['Rating', 'Player'], axis = 1, inplace = True)
df3.drop(['Rating', 'Player'], axis = 1, inplace = True)

df1.head()
```

```python
# 데이터 결합
new_df = pd.concat([df1, df2, df3], axis = 1)

new_df.head(20)
```

```python
new_df.isna().info()
```

```python
#  ▶ 3. 데이터 저장
new_df.to_csv('D:/Gray_Bigdata\src/00_Projects/data/spa_2021_tot_stats.csv')
```

# 통합 데이터 만들기(리그별)

```python
# ▶ 1. 데이터 불러오기
df1 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/20_21_players/Italy/ita_2021_tot_stats.csv', 
                  encoding = 'UTF8').drop(['Unnamed: 0'],axis=1)
df2 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/20_21_players/Italy/italy-serie-a-players-2020-to-2021-stats.csv', 
                  encoding = 'UTF8')
```
