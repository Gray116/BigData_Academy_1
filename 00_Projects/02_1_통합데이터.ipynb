```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# ▶ 1. 데이터 불러오기
df1 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/18_19_players/Spain/spa_1819_tot_stats.csv', 
                  encoding = 'UTF8')
df2 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/18_19_players/Spain/spain-la-liga-players-2018-to-2019-stats.csv', 
                  encoding = 'UTF8')
```

```python
df1.head()
```

```python
df1.drop(['Unnamed: 0'], axis = 1, inplace = True)

df1.head()
```

```python
df1 = df1.sort_values(by = ['Player'], ascending=True)

df1.reset_index
df1.head()
```

```python
df2.head()
```

```python
df2.drop(['birthday', 'birthday_GMT', 'league', 'season'], axis = 1, inplace = True)

df2.head()
```

```python
df2.rename(columns={'full_name' : 'Player'}, inplace = True)

df2.head()
```

```python
new_df.replace("", "0.0", inplace=True)
new_df.replace(np.nan, "0.0", inplace=True)

new_df.head(30)
```

```python
new_df.to_csv('D:/Gray_Bigdata/src/00_Projects/data/spa_1819_seanson.csv')
```

# 전체 데이터 결측치 확인 및 검사

```python
df_1 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/18_19_players/Total_Players/eng_1819_seanson.csv', 
                   encoding='UTF8')
df_2 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/18_19_players/Total_Players/fra_1819_seanson.csv', 
                  encoding='UTF8')
df_3 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/18_19_players/Total_Players/ger_1819_seanson.csv', 
                  encoding='UTF8')
df_4 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/18_19_players/Total_Players/ita_1819_seanson.csv', 
                  encoding='UTF8')
df_5 = pd.read_csv('D:/Gray_Bigdata/src/00_Projects/data/18_19_players/Total_Players/spa_1819_seanson.csv', 
                  encoding='UTF8')
```
