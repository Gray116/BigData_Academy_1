# 1절. 자연어 처리
- 자연어 : 사람들이 일상적으로 사용하는 언어<br>
- 자연어 처리 분야
    - 자연어 이해 : 형태소 분석 → 패턴 도출 → 의미 분석
    - 자연어 생성 : 대화 분석 → 다음에 나올 문장이나 단어 출력(딥러닝)
- 활용 분야 : 맞춤법검사, 번역기, 키워드 분석, 챗봇

# 2절. 자연어 처리 절차
- ① 전처리 : 단어, 어절 추출
- ② 분석 후보 생성 : 형태소 분리, 원형 복원, 품사 태깅
- ③ 제약조건(불용어 처리) 규칙 확인
- ④ 분석(연관분석, 거리계산, RNN, 현상시각화(워드클라우드), ...)

```python
# https://konlpy.org/en/latest/morph/#comparison-between-pos-tagging-classes
from konlpy.tag import Hannanum
han = Hannanum()
han.analyze(text) # 형태소 후보 반환
```

```python
# 형태소 분석만
print(han.morphs(text))
```

```python
# 명사만
print(han.nouns(text))
```

```python
print(han.pos(text, ntags = 9)) # 품사를 9가지로 나눔(default)
```

```python
print(han.pos(text, ntags = 22))
```

```python
# 퀴즈 : text중에서 형용사만 추출하기
tagged_text = han.pos(text, ntags = 22)
[ t[0] for t in tagged_text if (t[1] == 'PA') ]
```

```python
# 퀴즈 : text중에서 일반명사만 추출하기
[ t[0] for t in tagged_text if (t[1] == 'NC') ]
```

## 3.2 Kkma(꼬꼬마)

```python
text = """아름답지만 다소 복잡하기도 한 한국어는 전세계에서 13번째로 많이 사용되는 언어입니다"""
```

```python
from konlpy.tag import Kkma
kkma = Kkma(max_heap_size = 1024) # 힙 메모리 사이즈 늘릴 때 사용

print(kkma.morphs(text)) # 형태소 분석
```

```python
print(kkma.nouns(text)) # 명사 추출
```

```python
# pos tagging
print(kkma.pos(text)) # ntags = 30 (default)
```

```python
print(kkma.pos(text, 56))
```

```python
# 명사만 추출하기
tagged_text = kkma.pos(text, 56)
[ t[0] for t in tagged_text if t[1] == 'NNG']
```

# 7절. Word Embedding
- 단어 간 유사성 도출
- Word를 R차원의 Vector로 <b>매핑</b>시켜주는 것을 말한다.
    - 매핑 : 해당 값이 다른 값을 가리키도록 하는 것이다.

```python
# 뉴스 기사의 링크를 가져와서 웹 크롤링을 한다.
# 명사 추출 → 단어 간 거리의 유사성을 파악 및 출력 
```

```python
import requests
from bs4 import BeautifulSoup
rss_url = 'https://fs.jtbc.joins.com//RSS/economy.xml'

jtbc_economy = requests.get(rss_url)
soup = BeautifulSoup(jtbc_economy.content, 'xml')

linked_ele = soup.select('item > link')
link_list = [ link.text for link in linked_ele ]

link_list
```

```python
# 각 링크들의 뉴스 기사들에서 명사만 뽑아 단어들간의 거리를 계산하기
# [['명사1', '명사2', '명사3', ...], [], ...]
news = []

for link in link_list:
    response = requests.get(link)
    news_soup = BeautifulSoup(response.content, 'html.parser')
    
    news_title = news_soup.select_one(' div.title > h3#jtbcBody ').text # 전부 선택자가 같지 않을 수 있다!
    news_article = news_soup.select_one(' div#articlebody ').text
    
    nouns_list = kkma.nouns(news_title + news_article)
    news.append(nouns_list)    
```

```python
print('각 뉴스 당 명사의 갯수 :', [ len(n) for n in news ] )
```

```python
# pip install gensim 후 임베딩
from gensim.models import Word2Vec
```

```python
model = Word2Vec(news, window = 5,
                 min_count = 2, workers = -1)
```

```python
model.wv.most_similar('집', topn = 5) # 연관성 분석이 아님. '집'이라는 단어와 거리적(벡터)으로 가까운 단어들 출력
```
