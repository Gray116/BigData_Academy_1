```python
import numpy as np
x = np.arange(0, 10, 0.01)

x.shape
```

```python
plt.subplot(2, 1, 1) # 2행 1열의 subplot 생성하고, 그 중 n번째 subplot에 그래프를 그림
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 3) # 2행 2열의 subplot에서 세번째 subplot에 그래프를 그림
plt.plot(x, np.cos(x))

plt.subplot(2, 2, 4)
plt.plot(x, np.sin(x)*np.cos(x))

plt.show()
```

### 2.3.2 subplots() 함수로 서브플롯 그리기

```python
# 방법1 (2행 2열 서브플롯subplot)
plt.figure(figsize = (10, 5))
plt.subplot(2, 2, 1); plt.plot(x, np.sin(x))

plt.subplot(2, 2, 2); plt.plot(x, np.cos(x))

plt.subplot(2, 2, 3); plt.plot(x, np.tanh(x))

plt.subplot(2, 2, 4); plt.plot(x, np.sin(x)*np.cos(x))

plt.show()
```

```python
# 방법2 (subplots)
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 6))
axes[0, 0].plot(x, np.sin(x), color = 'r', linewidth = 3, linestyle = 'dashed')

axes[0, 1].plot(x, np.cos(x), 'g--') # 마커['c' = cyan]['--' = 점선]

axes[1, 0].plot(x, np.tanh(x), 'g')
axes[1, 0].plot([0, 1, 2, 3], [-1, -0.5, 0, 0.5], 'ko:') # k = 검정, o = 마커, : = 점선

axes[1, 1].plot(x, np.sin(x)*np.cos(x), 'r-.')
```

```python
# 방법3 (2행 2열 서브플롯)
# def sin_cos(x):
#     return np.sin(x)*np.cos(x)
# func_list = [np.sin, np.cos, np.tanh, sin_cos(x)]

func_list = [np.sin, np.cos, np.tanh, lambda x:np.sin(x)*np.cos(x)]
fmt_list = ['r', 'c--', 'g', 'r-.']

fig, axes = plt.subplots(2, 2, figsize = (10, 6))

for i, ax in enumerate(axes.flat):
    ax.plot(x, func_list[i](x), fmt_list[i])
```

```python
fig, axes = plt.subplots(ncols = 4, figsize = (18, 3), sharey = True) # nrows는 기본 값 1

for i, ax in enumerate(axes):
    ax.plot(x, func_list[i](x), fmt_list[i])
```

### 2.4.4 plt.scatter() 산점도 그래프

```python
np.random.seed(2021)

N = 50
x = np.random.rand(N) # 50개의 난수(실수)발생
y = np.random.rand(N)

colors = np.random.rand(N)
area = (np.random.rand(N)*30)**2

plt.scatter(x, y, 
            s = area,
            c = colors,
            alpha = 0.5) # s는 마커(점)들의 size
plt.show()
```

```python
plt.scatter(x = iris_df.petal_length, 
            y = iris_df.petal_width, 
            s = iris_df.petal_length * 10, 
            c = iris_df.petal_width * 20, alpha = 0.5)
plt.show()
```

## 2.5 다양한 그래프 그리기(2)
### 2.5.1 fill(), fill_between(), scatter()
### 2.5.2 bar(), barh(), axvline(), axhline()
### 2.5.3 hist(), boxplot(), violineplot()

```python
x = np.linspace(0, 10) # 0~10까지 일정한 간격으로 50개(기본 값) 수 추출
y = np.cos(x)

fig, axes = plt.subplots(2, 2, figsize = (10, 5))

axes[0, 0].scatter(x, y, s = 5, c = 'r', marker = 'v')
axes[0, 1].fill(x, y, c = 'b')
axes[1, 0].fill_between(x, y, color = 'r')
axes[1, 0].plot([0, 2], [-1, 1], c = 'k')
axes[1, 1].axvline(0.5)
axes[1, 1].axhline(0.7, c = 'r')

plt.show()
```

```python
fit, axes = plt.subplots(ncols=2)

axes[0].bar([1,2,3], [3,4,5], color = 'c')
axes[0].plot([1,2,3], [3,4,5], color = 'b')

axes[1].barh([1,2,3], [3,4,5], color = 'g')
axes[1].plot([1,2,3], [3,4,5], color = 'r')
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].hist(iris_df.sepal_length, bins = 20, color = 'r')
axes[0, 1].hist(iris_df.sepal_length, bins = 20, cumulative = True)
axes[1, 0].hist(iris_df.sepal_length, bins = 20, orientation = 'horizontal')
axes[1, 1].hist(iris_df.sepal_length, bins = 20, histtype = 'step')

plt.show()
```

## 2.6 그래프 커스터마이징
### 2.6.1 linestyle, linewidth
### 2.6.2 text(), annotate()

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)

fig, axes = plt.subplots(1, 2, figsize = (10,4)) # 1행 2열

axes[0].plot(x, y, linewidth = 5, linestyle = '--', color = 'purple')
axes[0].text(2, 0, '샘플 텍스트', fontsize = 14)
axes[1].scatter(x, y, marker = '*', color = 'r')
axes[1].annotate('Sine', xytext = (2, 0.75), # 글자가 있을 좌표 위치
                 xy = (5, 0.5), # 화살표가 가리킬 좌표 위치 
                 arrowprops = dict(arrowstyle = '->', 
                                   connectionstyle = 'angle3'))
plt.suptitle('메인 타이틀')
plt.show()
```

### 2.6.4 축과 눈금

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)

fig, axes = plt.subplots(2, 2, figsize = (10, 8))
plt.subplots_adjust(hspace = 0.4, wspace = 0.3) # 여백 조정
plt.suptitle('Main Title')

axes[0, 0].scatter(x, y, marker = '*')
axes[0, 0].set(title = 'An Example Axes', ylabel = 'y값', xlabel = 'x값', 
               xlim = [0, 12], ylim = [-2, 2])

axes[0, 1].scatter(x, y, color = 'r', marker = '+')
axes[0, 1].set_xlim(2, 6)
axes[0, 1].set_xlabel('x값', fontsize = 14)
axes[0, 1].set_ylim(-1.5, 1.5)

# 축 눈금 위치 조정
axes[1, 0].scatter(x, y, marker = '^', c = 'g')
axes[1, 0].set_xticks(range(0, 11, 3))
axes[1, 0].set_xticklabels(['0개', 3, '6개', 'max'])
axes[1, 0].set_yticks([-2, 0, 1, 2])
axes[1, 0].set_yticklabels(['min', '제로', 1, 'max'])

axes[1, 1].scatter(x, y, marker = '*', c = 'y')
axes[1, 1].set(xticks = [0, 2.5, 5, 7.5, 10], 
               xticklabels = ['Min', 2.5, 5, 7.5, 'Max'], 
               title = 'test')
axes[1, 1].grid(True)
axes[1, 1].spines['top'].set_visible(False)
axes[1, 1].spines['bottom'].set_position(('outward', 10))

plt.show()
```

### 2.6.5 축 공유

```python
x = np.arange(0, 10)
y1 = 0.5*x**2
y2 = -y1
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'g^:')
ax1.set(xlabel = 'X Data', ylabel = 'y1 data', title = '축 공유')

ax2 = ax1.twinx() # ax1과 x축을 공유
ax2.plot(x, y2, 'b*--')
ax2.set_ylabel('y2 data', color = 'b')

ax3 = ax1.twiny()
ax3.plot(-x, y1, 'ro-')
ax3.set_xlabel('-x data', color = 'r')

plt.show()
```
