### 2.2.2 연습2) 네이버 영화 랭킹 가져오기
```
9위 영화제목 - 영화소개 url
감독 및 배우
xxx : 감독 / xxx : 주연 / xxx : 주연
```

````python
# ▶ 영화순위, 영화제목, 소개url, 조연
import requests
from bs4 import BeautifulSoup
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'

movie_ranking = requests.get(url)
soup = BeautifulSoup(movie_ranking.content, 'html.parser')
title_list = soup.select('td.title > div.tit3 > a') # 영화제목

for idx in range(len(title_list)): # 영화순위 + 영화제목 출력하기
    link = url + title_list[idx].attrs['href']
    print("{:2d}위 영화 : {} - {}".format(idx+1, title_list[idx].text, link))
    
    page = requests.get(link) # link에서 주소를 얻어 page에 저장
    s = BeautifulSoup(page.content, "html.parser")
    
    people = s.select('li a.tx_people') # 사람이름
    staffs = s.select('dl.staff > dt') # 역할
    
    print('감독 및 배우')
    for idx in range(len(people)):
        print('{} : {}'.format(people[idx].text, staffs[idx].text), end = ' / ')
    print() # 자동 개행
````

````python
soup
````

````python
# import requests
# from bs4 import BeautifulSoup as bs
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'

movie_ranking = requests.get(url)
soup = bs(movie_ranking.content, 'html.parser')
title_list = soup.select('td.title > div.tit3 > a') # 영화제목

dict_list = []
for idx in range(len(title_list)): # 영화순위 + 영화제목 출력하기
    dic = {} # 빈 딕셔너리
    dic['rank'] = idx+1
    dic['title'] = title_list[idx].text
    dic['link'] = url + title_list[idx].attrs['href']
    
    page = requests.get(dic['link']) # link에서 주소를 얻어 page에 저장
    s = bs(page.content, "html.parser")
    people = s.select('li a.tx_people') # 사람이름
    staffs = s.select('dl.staff > dt') # 역할
    
    dic['staff'] = ''
    for idx in range(len(people)):
        dic['staff'] += "{} : {} / ".format(people[idx].text, staffs[idx].text)        
    dict_list.append(dic) # 자동 개행

dict_list
````

````python
import pandas as pd
df = pd.DataFrame(dict_list)
df.to_csv('Data/ch14_movie.csv')

df
````

https://goo.gl/HLntbd <br>
http://www.nextree.co.kr/p4327 <br><br>
정규표현식 연습장 : http://regexr.com<br><br>
간략 문법<br>
    **\d** : 숫자와 매치, [0-9]와 동일<br>
    **\D** : 숫자가 아닌 것<br>
    **\s** : whitespace<br>
    **\w** : 영문자나 숫자<br>
    **\W** : 영문자나 숫자가 아닌 문자<br>
    **.** : 문자 한개<br>
    **+** : 한번 이상 반복<br>
    * : 0번 이상 반복<br>
    **{2,4}** : 2~4회 반복<br>
    **?** : {0,1}의 의미<br>
    **^** : 시작<br>
    $ : 끝

### 2.3.1 연습1) 전국 날씨 RSS 데이터 가져오기

````python
import re, urllib.request
url = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
target = urllib.request.urlopen(url)
contents = target.read().decode('UTF8')
locations = re.findall(r'<location wl_ver="3">.+?</location>', contents, 
                       re.DOTALL)

for loc in locations:
    province = re.search(r'<province>(.+)</province>', loc)
    city = re.search(r'<city>(.+)</city>', loc)
    print("\n" + province.group(1), "지역의 ▶", city.group(1), '◀ 날씨입니다.')
    
    data = re.findall(r'<data>.+?</data>', loc, re.DOTALL)
    
    ampm = '오전'
    for item in data:
        tmEf = re.search(r'<tmEf>(.+)</tmEf>', item).group(1) # 날짜 및 시간
        wf = re.search(r'<wf>(.+)</wf>', item).group(1) # 날씨
        tmn = re.search(r'<tmn>(.+)</tmn>', item).group(1) # 최저 기온
        tmx = re.search(r'<tmx>(.+)</tmx>', item).group(1) # 최고 기온
        
        print('{} : {}\n날씨 : {}\n최저 : {}도\n최고 : {}도'.\
              format(tmEf[:10], ampm, wf, tmn, tmx) )
        
        if ampm == '오전':
            ampm = '오후'
        else:
            ampm = '오전'
        
````

## 2.4 BeautifulSoup을 이용한 XML 크롤링

````python
url = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'

import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs

# target = requests.get(url)
# soup = bs(target.content, "xml")
target = urlopen(url)
soup = bs(target, "xml")

locations = soup.find_all('location')
# locations = soup.select('location')

for loc in locations:
    province = loc.select_one('province').text # find 사용가능
    city = loc.select_one('city').text
    data = loc.select('data') # find_all 사용가능
    
    print('\n' + province, '지역의 ▶', city, '◀ 도시 날씨입니다.')
    
    ampm = '오전'
    for item in data:
        tmEf = item.find('tmEf').text[:10]
        wf = item.find('wf').text
        tmn = item.find('tmn').text
        tmx = item.find('tmx').text
        print('{} {}\n날씨 : {}\n최저 : {}도\n최고 : {}도'.\
              format(tmEf, ampm, wf, tmn, tmx))
        
        if ampm == '오전':
            ampm = '오후'
        else:
            ampm = '오전'    
````
