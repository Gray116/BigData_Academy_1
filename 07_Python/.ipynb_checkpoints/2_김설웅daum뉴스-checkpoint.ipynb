```python
import requests
from bs4 import BeautifulSoup as bs
url = 'https://news.daum.net/ranking/popular'
news_idx = requests.get(url)
soup = bs(news_idx.content, "html.parser")

newstitle = soup.select('ul.list_news2 > li > div.cont_thumb > strong.tit_thumb > a.link_txt')
newscomname = soup.select('div.cont_thumb > strong.tit_thumb > span.info_news')

dict_list = []
for idx in range(len(newstitle)):
    dic = {}
    
    dic['번호'] = idx+1
    dic['뉴스 제목'] = newstitle[idx].text
    dic['신문사명'] = newscomname[idx].text
    
    dict_list.append(dic)

dict_list
```

```python
import pandas as pd
news_df = pd.DataFrame(dict_list)
news_df.to_csv('D:/Gray_Bigdata/src/07_Python/2_news.csv')

news_df
```
