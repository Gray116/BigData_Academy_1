```python
import requests
from bs4 import BeautifulSoup as bs
url = 'http://www.yes24.com/24/category/bestseller'
book_idx = requests.get(url)
soup = bs(book_idx.content, "html.parser")

book_title = soup.select('li > p.copy + p + p > a')
book_aupu = soup.select('li > p.aupu')
book_price = soup.select('li > p.price > strong')

dict_list = []
for idx in range(len(book_title)):
    dic = {}
    
    dic['번호'] = idx+1
    dic['책 제목']= book_title[idx].text
    dic['저자 및 출판사'] = book_aupu[idx].text
    dic['가격'] = book_price[idx].text
    
    dict_list.append(dic)
    
dict_list
```

```python
import pandas as pd
book_df = pd.DataFrame(dict_list)
book_df.to_csv('D:/Gray_Bigdata/src/07_Python/1_yes24.csv', encoding='CP949')

book_df
```
